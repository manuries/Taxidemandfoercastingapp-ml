```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# 1. Load the trimmed dataset
df = pd.read_csv("yellow_tripdata_small.csv")

# 2. Convert pickup datetime to proper format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])

# 3. Extract hour of day
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour

# 4. Aggregate demand by hour
demand_by_hour = df.groupby('pickup_hour').size().reindex(range(24), fill_value=0).reset_index(name='trip_count')

# 5. Train a simple regression model
X = demand_by_hour[['pickup_hour']]
y = demand_by_hour['trip_count']

model = LinearRegression()
model.fit(X, y)

# 6. Predict demand for all 24 hours
all_hours = pd.DataFrame({'pickup_hour': range(24)})
all_pred = model.predict(all_hours)

# 7. Evaluate model
y_pred = model.predict(X)
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)

# 8. Plot actual vs predicted demand
plt.figure(figsize=(12,6))
plt.bar(demand_by_hour['pickup_hour'], demand_by_hour['trip_count'], alpha=0.6, label='Actual Demand')
plt.plot(all_hours['pickup_hour'], all_pred, color='red', marker='o', label='Predicted Demand')
plt.xticks(range(24))
plt.xlabel("Hour of Day (0=Midnight, 23=11PM)")
plt.ylabel("Trip Count")
plt.title("Taxi Demand Forecasting by Hour")
plt.legend()
plt.show()
```
